```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Produção agrícola Brasil x RS 2022

```python
dfbrasil = pd.read_csv('ranking-agricultura-valor-da-produo-brasil.csv')
```

```python
dfbrasil
```

```python
dfrs = pd.read_csv('ranking-agricultura-valor-da-produo-rio-grande-do-sul.csv')
```

```python
dfrs
```

```python
dfbrasil = dfbrasil.dropna()
```

```python
dfbrasil
```

```python
dfrs = dfrs.dropna()
```

```python
dfrs
```

```python
dfrs = dfrs.rename(columns ={'Produto RS': 'Produto', 'valor': 'Valor produzido RS'})
```

```python
dfrs
```

```python
dfbrasil = dfbrasil.rename(columns ={'Produto Brasil': 'Produto', 'Valor produzido': 'Valor produzido BR'})
```

```python
dfbrasil
```

```python
dfbrasil = dfbrasil.drop(columns={'unidade'})
```

```python
dfbrasil
```

```python
dfbrasil['valor'] = dfbrasil['valor'].str.replace(r'[^\d\.\-]', '', regex=True).str.strip()
```

```python
dfbrasil['valor']
```

```python
dfbrasil['valor'][0]
```

```python
dfbrasil['valor'][0].replace('.','')
```

```python
dfbrasil = pd.DataFrame(dfbrasil)
```

```python
print(dfbrasil.dtypes)
```

```python
dfbrasil
```

```python
dfcombinado = pd.merge(dfbrasil, dfrs, how='inner')
```

```python
dfcombinado
```

```python
dfcombinado = dfcombinado.rename(columns ={'valor': 'Valor Produzido BR'})
```

```python
dfcombinado
```

```python
dfcombinado = dfcombinado.drop(columns={'unidade'})
```

```python
dfcombinado
```

```python
dfcombinado['Valor Produzido BR'] = dfcombinado['Valor Produzido BR'].str.replace('.', '').astype(str)
dfcombinado['Valor produzido RS'] = dfcombinado['Valor produzido RS'].str.replace('.', '').astype(str)
```

```python
dfcombinado
```

```python
dfcombinado['Valor Produzido BR'] = dfcombinado['Valor Produzido BR'].astype(int)
dfcombinado['Valor produzido RS'] = dfcombinado['Valor produzido RS'].astype(int)
```

```python
dfcombinado
```

```python
dfcombinado['Diferença'] = dfcombinado['Valor Produzido BR'] - dfcombinado['Valor produzido RS']
```

```python
dfcombinado
```

```python
dfcombinado['Quantidade RS (%)'] = (dfcombinado['Valor produzido RS'] / dfcombinado['Valor Produzido BR']) * 100

```

```python
dfcombinado
```

```python
dfcombinado['Quantidade RS (%)'] = dfcombinado['Quantidade RS (%)'].round(2)
dfcombinado
```

```python
dfcombinado['Quantidade RS (%)'] = pd.to_numeric(dfcombinado['Quantidade RS (%)'], errors='coerce')
```

```python
dfcombinado
```

```python
df_classificado = dfcombinado.sort_values(by='Quantidade RS (%)', ascending=False)
```

```python
df_classificado
```

```python
top_10 = df_classificado.head(10)
```

```python
top_10
```

```python
import plotly.express as px
```

```python
fig = px.bar(top_10, x='produto', y='Quantidade RS (%)', title='Top 10 Maiores Quantidades Produzidas pelo RS em relação ao Brasil')
fig.update_traces(marker_color='rgb(200,130,100)', marker_line_color='rgb(8,48,107)', marker_line_width=1.5, opacity=0.6)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
top_10_menores = df_classificado.tail(10)
```

```python
top_10_menores
```

```python
fig = px.bar(top_10_menores, x='produto', y='Quantidade RS (%)', title='Top 10 Maiores Quantidades Produzidas pelo RS em relação ao Brasil')
fig.update_traces(marker_color='rgb(200,130,100)', marker_line_color='rgb(8,48,107)', marker_line_width=1.5, opacity=0.6)
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
